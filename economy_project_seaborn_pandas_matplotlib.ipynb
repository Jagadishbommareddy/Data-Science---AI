# distplot
- distribution -- we understand in stats part later 
- uniform distribution 
- normal distribuion = gausian distribution = bell curve 
- probability distribution ( image classificaiton, text classification algorithm)


```python
#Distributions
vs1 = sns.distplot(stats['InternetUsers']) # This is UNIVARIATE
vs1
```

# bins
The bins parameter enables you to control the bins of the histogram (i.e., the number of bars)

```python

vs1 = sns.distplot(stats['InternetUsers'],bins=20) # This is UNIVARIATE
vs1
```

# Box PLOTS
- A box plot is a graph summarising a set of data. 
- The shape of the boxplot shows how the data is distributed and it also shows any outliers

# Outlier
- outlier is the datapoint which is very far from other observation 
- statistic OUTLIER == ANOMALY DETECTION 

```python
#Boxplot
vs2 = sns.boxplot(data=stats,x='IncomeGroup',y='BirthRate') # BIVARIANT
vs2
```

# lmplot
- lmplot ( linear model plot) -- linear regression 


```python
vs3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False)#lm - linear model
vs3
```

```python
vs4 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate')#lm - linear model
vs4
```

```python
vis5 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
vis5
```
